```python
import pandas as pd
df = pd.read_csv('apple.csv', index_col='Date', parse_dates=True)
df = df.sort_index()
print(df.info())
```

```python
df.loc['2012-Feb', 'Close'].mean()
```

```python
df.loc['2012-Feb':'2015-Feb', 'Close'].mean()
```

```python
df.resample('W')['Close'].mean()
```

```python
import matplotlib.pyplot as plt

new_sample_df = df.loc['2012-Feb':'2017-Feb', ['Close']]
new_sample_df.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [25, 32, 34, 20, 25]

plt.title("График функции y", color='#173554', fontstyle="italic", fontweight="bold")
plt.xlabel('Ось х', color='#173554', fontweight="bold") #Подпись для оси х
plt.ylabel('Ось y', color='#173554', fontweight="bold")
ax = plt.gca()
ax.set_facecolor('#84d8d8')
plt.gcf().set_facecolor('#84d8d8')


plt.plot(x, y, color='#173554', marker='o', markersize=7)
plt.show()
```

```python

new_sample_df = df.loc['2012-Feb':'2017-Feb', ['Close']]
new_sample_df.plot()

plt.title("Стоимость акций apple", color='#173554', fontstyle="italic", fontweight="bold")
plt.xlabel('Дата', color='#173554', fontweight="bold") #Подпись для оси х
plt.ylabel('Стоимость', color='#173554', fontweight="bold")
ax = plt.gca()
ax.set_facecolor('#84d8d8')
plt.gcf().set_facecolor('#84d8d8')
plt.legend(facecolor='#84d8d8', edgecolor='#173554')


plt.plot(color='#173554')
plt.show()
```
